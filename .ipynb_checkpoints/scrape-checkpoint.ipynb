```python
import wikipedia
import datetime
import pandas as pd
```

```python
page=wikipedia.page(title="List of suicides")
```

```python
text=page.content.split('\n')
```

```python
df=pd.DataFrame(text)
df=df[df[0] != ''].loc[4:]
df['name(year)'], df['info'] = df[0].str.split(',', 1).str
df['name'],df['year'] = df['name(year)'].str.split('(', 1).str
df['year']=df['year'].str.rstrip(')')
df=df[['name','year','info']].dropna()
```
